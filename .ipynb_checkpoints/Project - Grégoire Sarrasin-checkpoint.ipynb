##### Topics

The idea is to structure and qualitatively describe the Swiss health insurance market. And finally to make prediction on future individual premiums, or at least on cost evolution.

Some definitions:
* Health cost = sum of the medicals bills, by canton/age class.
* Premium = health cost + administrative cost +/- financial reserves, by canton/age class, divided per insured.
* Ordinary premium = standard model, franchise Fr. 300.- with accident risk (per age class/canton/insurer)
* Average premium = ordinary premium per age class/canton

Concerning the premiums, they are about 250'000 different per year on the whole country. Premiums change regarding the age class, the insurer, the franchise level, the insurance model, the cost area (1 to 3 per canton) and the accident inclusion. But all start from the ordinary premium, others premiums are calculated with discounts ranges defined by the law. Average premium remove the "insurer effect". The idea is that all the premiums are declinations of a reference premium, which implies that patterns and relationships in the data can be studied at an aggregate of this level, or one premium per canton.

The analysis of data will be used to determine how components of costs and external components involves these costs, including through years. Moreover, a description of the insurance market will be done: structure of the subpolts of insured people (per canton and per gender) and relative trends.

Predicting will be used to determine premiums on the 2017 exercice, the last year with available data.

##### Features analysed

The following features per canton from the "Portal Statistics on compulsory health insurance" of the Federal Office of Public Health (FOPH) are included in the model:

- providers density [density = per inhabitant] - 9.01 and external source for "drogueries" *
- group costs - 2.26 **
- insured population and gender - 7.03
- premiums - 8.02 and 8.04
- subsidy - extract from 4.02, 4.07 and 4.08

The following features per canton from the population census are included too:
- aging rate [= rate of people over 65 yo]
- child rate [= rate of poeple below 19 yo]
- urbanisation rate
- birth/death/marriage/divorce rates ***
- foreign national rate
- percentage of language spoken
- religious orientation
- wooded/built/agricultural/unproductive area
- etc.

Many others features related by health market may be included in the models, but some are really hard to obtain and to compute. For example, there is the unit price of the medical cares which vary over the years and the insurers. They are negociations between insurers and medical providers and they are regularly challenged, involving retroactive modifications, sometimes anticipated by the insurers when fixing the premiums. Another example is the evolution of technology and science. Health staff do their job faster and more efficiently for the same cost-to-the-act and on another hand, more specific drugs are discovered at a cost always rising. Finally, there are regular legal modifications which shift costs to the basis insurance like the new hospital funding (2012, shifting of costs from the complementary insurance to the basis one) or the rules in case of pregnancy (2014, shifting of costs from the insured's participation to the basis insurance).

The idea is to target features known at a large scale, like providers density, aging population, urbanisation which can all give result in the frame of this work. Other features are inculded on a prospective way, or in view to identify "way of life" implying a premium cost (BS is 69% more expensive than AI): forest are known to improve people's health, can we see it in the premiums? And unemployment or religion? Politic parties have some values (liberal and economics for FDP, family and religion for CVP, tradition and patriotism for SVP, ...) and the assumption of a correlation between this values and the voting people's values implies a lecture grid of the population's values and it is why this feature is part of the model.

#### Three axis analysis

The difficulty with merging tables from different sources is to deal with the particular limitations. In other words, linking data is done at the level of the weakest. More precisely here, we have temporal statistics and the shortest table will determine the final time boundaries of the data.

This is why I choose to separate the work in three parts, respectively one per source, each part being construct on the basis of the preceding:
- part One, source from Federal Office of Public Health (FOPH): this is the most aboundant source, which allow to cover years from 2004 to 2017 (and life expectancy from FSO which is available since 1982 *)
- part Two, source from the Federal Statistical Office (FSO): the census population data is used here and due to a change in collect method, yearly data are only available since 2010
- part Three, source from santésuisse, the health insurers' lobby with the detail of the health costs per provider

\* FSO sent me this data on request, which wasn't available online.

```python
## organize and summarize the data ##

# We want to approach the whole market of health insurance, it's why we will use data from the Federal
# Statistic Office, Federal Office of Public Health and santésuiss (health insurers lobby).
# Data are available in the 'data' folder and we import it here, first separately from the census because
# census data are yearly available from 2010 and selected statistics on health insurance are continuous 
# from 2004.

# first step: importing and structuring data

import pandas as pd
import numpy as np
import os

def format_density(file): # clean input data
    if(file.shape[1] != 9): # data before 2009 has two more columns and since 2009 are 9 columns big
        file.drop(file.columns[10:], axis=1, inplace=True)
        file.drop(file.columns[6], axis=1, inplace=True)
    file.columns = ['Canton', 'Mean_living_population', 'General_practitioners', 
                        'Specialised_doctors', 'Total_doctors', 'Doctors_density', 
                        'Pharmacies', 'Type_of_drugs_providing', 'Pharmacies_density']
    file.drop(file.index[0:5], inplace=True)
    file.drop(file.index[26:], inplace=True)
    file.reset_index(drop=True, inplace=True)
    file.insert(0, 'Year', -1)
    return file

# providers density 2004-2017
density2004_data = format_density(pd.read_excel(os.path.join('data', 'density_data', '2004.xls'))).replace({'Year': -1}, 2004)
density2005_data = format_density(pd.read_excel(os.path.join('data', 'density_data', '2005.xls'))).replace({'Year': -1}, 2005)
density2006_data = format_density(pd.read_excel(os.path.join('data', 'density_data', '2006.xls'))).replace({'Year': -1}, 2006)
density2007_data = format_density(pd.read_excel(os.path.join('data', 'density_data', '2007.xls'))).replace({'Year': -1}, 2007)
density2008_data = format_density(pd.read_excel(os.path.join('data', 'density_data', '2008.xls'))).replace({'Year': -1}, 2008)
density2009_data = format_density(pd.read_excel(os.path.join('data', 'density_data', '2009.xls'))).replace({'Year': -1}, 2009)
density2010_data = format_density(pd.read_excel(os.path.join('data', 'density_data', '2010.xls'))).replace({'Year': -1}, 2010)
density2011_data = format_density(pd.read_excel(os.path.join('data', 'density_data', '2011.xls'))).replace({'Year': -1}, 2011)
density2012_data = format_density(pd.read_excel(os.path.join('data', 'density_data', '2012.xls'))).replace({'Year': -1}, 2012)
density2013_data = format_density(pd.read_excel(os.path.join('data', 'density_data', '2013.xls'))).replace({'Year': -1}, 2013)
density2014_data = format_density(pd.read_excel(os.path.join('data', 'density_data', '2014.xls'))).replace({'Year': -1}, 2014)
density2015_data = format_density(pd.read_excel(os.path.join('data', 'density_data', '2015.xls'))).replace({'Year': -1}, 2015)
density2016_data = format_density(pd.read_excel(os.path.join('data', 'density_data', '2016.xls'))).replace({'Year': -1}, 2016)
density2017_data = format_density(pd.read_excel(os.path.join('data', 'density_data', '2017.xls'))).replace({'Year': -1}, 2017)

def format_population(file): # clean input data
    file.dropna(inplace=True)
    file.drop(file.columns[5:], axis=1, inplace=True)
    file.columns = ['Canton', 'Men', 'Women', 'Children', 'Total_population']
    file.drop(file.index[0], inplace=True)
    file.drop(file.index[26:], inplace=True)
    file.reset_index(drop=True, inplace=True)
    file.insert(0, 'Year', -1)
    file.Men = file.Men / file.Total_population # to make cantons comparable
    file.Women = file.Women / file.Total_population
    file.Children = file.Children / file.Total_population
    return file


# population and gender 2004-2017
population2004_data = format_population(pd.read_excel(os.path.join('data', 'population_data', '2004.xls'))).replace({'Year': -1}, 2004)
population2005_data = format_population(pd.read_excel(os.path.join('data', 'population_data', '2005.xls'))).replace({'Year': -1}, 2005)
population2006_data = format_population(pd.read_excel(os.path.join('data', 'population_data', '2006.xls'))).replace({'Year': -1}, 2006)
population2007_data = format_population(pd.read_excel(os.path.join('data', 'population_data', '2007.xls'))).replace({'Year': -1}, 2007)
population2008_data = format_population(pd.read_excel(os.path.join('data', 'population_data', '2008.xls'))).replace({'Year': -1}, 2008)
population2009_data = format_population(pd.read_excel(os.path.join('data', 'population_data', '2009.xls'))).replace({'Year': -1}, 2009)
population2010_data = format_population(pd.read_excel(os.path.join('data', 'population_data', '2010.xls'))).replace({'Year': -1}, 2010)
population2011_data = format_population(pd.read_excel(os.path.join('data', 'population_data', '2011.xls'))).replace({'Year': -1}, 2011)
population2012_data = format_population(pd.read_excel(os.path.join('data', 'population_data', '2012.xls'))).replace({'Year': -1}, 2012)
population2013_data = format_population(pd.read_excel(os.path.join('data', 'population_data', '2013.xls'))).replace({'Year': -1}, 2013)
population2014_data = format_population(pd.read_excel(os.path.join('data', 'population_data', '2014.xls'))).replace({'Year': -1}, 2014)
population2015_data = format_population(pd.read_excel(os.path.join('data', 'population_data', '2015.xls'))).replace({'Year': -1}, 2015)
population2016_data = format_population(pd.read_excel(os.path.join('data', 'population_data', '2016.xls'))).replace({'Year': -1}, 2016)
population2017_data = format_population(pd.read_excel(os.path.join('data', 'population_data', '2017.xls'))).replace({'Year': -1}, 2017)

def format_premiums(file): # clean input data
    file.drop(file.columns[6:], axis=1, inplace=True)
    file.drop(file.columns[1:5], axis=1, inplace=True)
    file.columns = ['Canton', 'Premium']
    file.dropna(inplace=True)
    file.drop(file.index[0], inplace=True)
    file.reset_index(drop=True, inplace=True)
    file.insert(0, 'Year', -1)
    file.drop(file.index[26:], inplace=True)
    return file


# premiums adults 2004-2017
premiums2004_data = format_premiums(pd.read_excel(os.path.join('data', 'premiums_data', '2004A.xls'))).replace({'Year': -1}, 2004)
premiums2005_data = format_premiums(pd.read_excel(os.path.join('data', 'premiums_data', '2005A.xls'))).replace({'Year': -1}, 2005)
premiums2006_data = format_premiums(pd.read_excel(os.path.join('data', 'premiums_data', '2006A.xls'))).replace({'Year': -1}, 2006)
premiums2007_data = format_premiums(pd.read_excel(os.path.join('data', 'premiums_data', '2007A.xls'))).replace({'Year': -1}, 2007)
premiums2008_data = format_premiums(pd.read_excel(os.path.join('data', 'premiums_data', '2008A.xls'))).replace({'Year': -1}, 2008)
premiums2009_data = format_premiums(pd.read_excel(os.path.join('data', 'premiums_data', '2009A.xls'))).replace({'Year': -1}, 2009)
premiums2010_data = format_premiums(pd.read_excel(os.path.join('data', 'premiums_data', '2010A.xls'))).replace({'Year': -1}, 2010)
premiums2011_data = format_premiums(pd.read_excel(os.path.join('data', 'premiums_data', '2011A.xls'))).replace({'Year': -1}, 2011)
premiums2012_data = format_premiums(pd.read_excel(os.path.join('data', 'premiums_data', '2012A.xls'))).replace({'Year': -1}, 2012)
premiums2013_data = format_premiums(pd.read_excel(os.path.join('data', 'premiums_data', '2013A.xls'))).replace({'Year': -1}, 2013)
premiums2014_data = format_premiums(pd.read_excel(os.path.join('data', 'premiums_data', '2014A.xls'))).replace({'Year': -1}, 2014)
premiums2015_data = format_premiums(pd.read_excel(os.path.join('data', 'premiums_data', '2015A.xls'))).replace({'Year': -1}, 2015)
premiums2016_data = format_premiums(pd.read_excel(os.path.join('data', 'premiums_data', '2016A.xls'))).replace({'Year': -1}, 2016)
premiums2017_data = format_premiums(pd.read_excel(os.path.join('data', 'premiums_data', '2017A.xls'))).replace({'Year': -1}, 2017)

def format_subsidy(file): # clean input data
    file.insert(0, 'Year', -1)
    file.replace('-', 0, inplace=True)
    return file

# subsidy data 2004-2017
subsidy2004_data = format_subsidy(pd.read_excel(os.path.join('data', 'subsidy_data', '2004.xls'))).replace({'Year': -1}, 2004)
subsidy2005_data = format_subsidy(pd.read_excel(os.path.join('data', 'subsidy_data', '2005.xls'))).replace({'Year': -1}, 2005)
subsidy2006_data = format_subsidy(pd.read_excel(os.path.join('data', 'subsidy_data', '2006.xls'))).replace({'Year': -1}, 2006)
subsidy2007_data = format_subsidy(pd.read_excel(os.path.join('data', 'subsidy_data', '2007.xls'))).replace({'Year': -1}, 2007)
subsidy2008_data = format_subsidy(pd.read_excel(os.path.join('data', 'subsidy_data', '2008.xls'))).replace({'Year': -1}, 2008)
subsidy2009_data = format_subsidy(pd.read_excel(os.path.join('data', 'subsidy_data', '2009.xls'))).replace({'Year': -1}, 2009)
subsidy2010_data = format_subsidy(pd.read_excel(os.path.join('data', 'subsidy_data', '2010.xls'))).replace({'Year': -1}, 2010)
subsidy2011_data = format_subsidy(pd.read_excel(os.path.join('data', 'subsidy_data', '2011.xls'))).replace({'Year': -1}, 2011)
subsidy2012_data = format_subsidy(pd.read_excel(os.path.join('data', 'subsidy_data', '2012.xls'))).replace({'Year': -1}, 2012)
subsidy2013_data = format_subsidy(pd.read_excel(os.path.join('data', 'subsidy_data', '2013.xls'))).replace({'Year': -1}, 2013)
subsidy2014_data = format_subsidy(pd.read_excel(os.path.join('data', 'subsidy_data', '2014.xls'))).replace({'Year': -1}, 2014)
subsidy2015_data = format_subsidy(pd.read_excel(os.path.join('data', 'subsidy_data', '2015.xls'))).replace({'Year': -1}, 2015)
subsidy2016_data = format_subsidy(pd.read_excel(os.path.join('data', 'subsidy_data', '2016.xls'))).replace({'Year': -1}, 2016)
subsidy2017_data = format_subsidy(pd.read_excel(os.path.join('data', 'subsidy_data', '2017.xls'))).replace({'Year': -1}, 2017)

def format_expectancy(file): # clean input data
    file.columns = file.iloc[1]
    file.drop(file.index[:3], inplace=True)
    file.drop(file.index[27:], inplace=True)
    file = file.melt(id_vars=['Canton'], var_name='Year', value_name='Expectancy')
    return file

# life expectancy 1982-2017
expectancy_men = format_expectancy(pd.read_excel(os.path.join('data', 'expectancy_data', 'life_expectancy_82-17.xlsx'), sheet_name='EV 0 hommes'))
expectancy_women = format_expectancy(pd.read_excel(os.path.join('data', 'expectancy_data', 'life_expectancy_82-17.xlsx'), sheet_name='EV 0 femmes'))
expectancy_men_65 = format_expectancy(pd.read_excel(os.path.join('data', 'expectancy_data', 'life_expectancy_82-17.xlsx'), sheet_name='EV 65 hommes'))
expectancy_women_65 = format_expectancy(pd.read_excel(os.path.join('data', 'expectancy_data', 'life_expectancy_82-17.xlsx'), sheet_name='EV 65 femmes'))
```

```python
## second step: merging the tables in one

densities = [density2004_data, density2005_data, density2006_data, density2007_data, density2008_data, 
             density2009_data, density2010_data, density2011_data, density2012_data, density2013_data, 
             density2014_data, density2015_data, density2016_data, density2017_data]
populations = [population2004_data, population2005_data, population2006_data, population2007_data, 
              population2008_data, population2009_data, population2010_data, population2011_data, 
              population2012_data, population2013_data, population2014_data, population2015_data, 
              population2016_data, population2017_data]
subsidies = [subsidy2004_data, subsidy2005_data, subsidy2006_data, subsidy2007_data, subsidy2008_data, 
            subsidy2009_data, subsidy2010_data, subsidy2011_data, subsidy2012_data, subsidy2013_data, 
            subsidy2014_data, subsidy2015_data, subsidy2016_data, subsidy2017_data]
premiums = [premiums2004_data, premiums2005_data, premiums2006_data, premiums2007_data, premiums2008_data, 
           premiums2009_data, premiums2010_data, premiums2011_data, premiums2012_data, premiums2013_data, 
           premiums2014_data, premiums2015_data, premiums2016_data, premiums2017_data]

aggregated_data = []

for i in range(len(densities)):
    if(i<1):
        result = pd.merge(densities[i], populations[i], on=['Year', 'Canton'])
        result = pd.merge(result, subsidies[i], on=['Year', 'Canton'])
        aggregated_data = pd.merge(result, premiums[i], on=['Canton', 'Year'])
    else:
        result = pd.merge(densities[i], populations[i], on=['Year', 'Canton'])
        result = pd.merge(result, subsidies[i], on=['Year', 'Canton'])
        result = pd.merge(result, premiums[i], on=['Canton', 'Year'])
        aggregated_data = pd.concat([aggregated_data, result])

aggregated_data = pd.merge(aggregated_data, expectancy_men, on=['Year', 'Canton'])
aggregated_data = pd.merge(aggregated_data, expectancy_women, on=['Year', 'Canton']).rename(columns = {'Expectancy_x' : 'Expectancy Men'})
aggregated_data = pd.merge(aggregated_data, expectancy_men_65, on=['Year', 'Canton']).rename(columns = {'Expectancy_y' : 'Expectancy Women'})
aggregated_data = pd.merge(aggregated_data, expectancy_women_65, on=['Year', 'Canton']).rename(columns = {'Expectancy_x' : 'Expectancy Men 65'})
aggregated_data = aggregated_data.rename(columns = {'Expectancy_y' : 'Expectancy Women 65'}) 
    # renaming columns is shifted because applied on the already merged data
```

```python
# now we have an aggregated table which need to be refined.

aggregated_data.Premium = 12*aggregated_data.Premium # change monthly premium in annualy premium

aggregated_data.insert(aggregated_data.shape[1] -5, 'Subsidy_ratio', # additional feature
                       aggregated_data['Subsidy per person'] / aggregated_data['Premium']) 

# creating % of population or densities to make cantons comparable
aggregated_data.insert(aggregated_data.shape[1] -8, 'Complementary_benefits_ratio',
                      aggregated_data['Nb of people with PC'] / aggregated_data['Total_population'])

aggregated_data.insert(aggregated_data.shape[1] -8, 'Welfare_ratio',
                      aggregated_data['Nb of people with social help'] / aggregated_data['Total_population'])

aggregated_data.insert(2, 'General practitioners density', 
                       1000 * aggregated_data['General_practitioners'] / aggregated_data['Total_population'])

aggregated_data.insert(3, 'Specialized doctors density', 
                      1000 * aggregated_data['Specialised_doctors'] / aggregated_data['Total_population'])

aggregated_data.drop(aggregated_data.columns[4], axis=1, inplace=True) # redundant with Total population

aggregated_data.drop(aggregated_data.columns[6], axis=1, inplace=True) 
                            # linear combination of General and Specialized

aggregated_data.reset_index(drop=True, inplace=True)
```

```python
# note about the doctors densities: It is commonly accepted that "supply creates demand" in particular 
# for doctors because they are gateways in health cares. But it won't appears in analysis clearly. Why?
# The counting methodology changed, in particular for general practitioners. Since 2007, specialists 
# in internal medicine are considered as general practitioners. Note from the FOPH: "Inklusiv Fachärzte 
# für innere Medizin, die als Grundversorger tätig sind (ab 2007)." This mean the data for doctors are 
# not usable before 2007 in our model.
# Concerning the total doctors density, another FOPH note explain the 2008 break: "Ab 2008 ohne die 
# Ärzte, die weder im ambulanten noch im stationären Sektor arbeiten (Versicherungen, Bundesverwaltung 
# usw.)." (since 2008 are not counted doctors who don't work in the ambulatory or stationary sector).
# Due to these breaks in the data, we will put the detailed densities aside for the moment.

plt.figure(figsize=(17,5))
plt.subplot(1, 3, 1)
plt.scatter(data['Year'], data['General practitioners density'])
plt.title('Density of general practitioners in Switzerland, per canton')
plt.xlabel('Year')
plt.ylabel('General practitioners density')

plt.subplot(1, 3, 2)
plt.scatter(data['Year'], data['Specialized doctors density'])
plt.xlabel('Year')
plt.ylabel('Specialized doctors density')

plt.subplot(1, 3, 3)
plt.scatter(data['Year'], data['Doctors_density'])
plt.xlabel('Year')
plt.ylabel('Doctors density')

plt.show()
```

```python
# let's apply PCA (Principal components analysis) which indicates the features which explain the best 
# the target by dimensionality reduction:

X = data_df.drop(['Premium'], axis=1).values
y = data_df.Premium.values

print(X.shape)
print(y.shape)

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_rescaled = scaler.fit_transform(X)

pca = PCA(n_components=2)

X_2d_rescaled = pca.fit_transform(X_rescaled)
plt.figure(figsize=(18,10))

for canton in data_df.iloc[:, 19:]:
    idx = (data_df.loc[:, canton] == 1)

    # Plot their components
    if(canton == 'Canton_BS'):
        plt.scatter(X_2d_rescaled[idx, 0], X_2d_rescaled[idx, 1], label='{}'.format(canton), marker='^')
    elif(canton == 'Canton_GE'):
        plt.scatter(X_2d_rescaled[idx, 0], X_2d_rescaled[idx, 1], label='{}'.format(canton), marker='2')
    elif(canton == 'Canton_NW'):
        plt.scatter(X_2d_rescaled[idx, 0], X_2d_rescaled[idx, 1], label='{}'.format(canton), marker='1')
    elif(canton == 'Canton_AI'):
        plt.scatter(X_2d_rescaled[idx, 0], X_2d_rescaled[idx, 1], label='{}'.format(canton), marker='v')
    else:
        plt.scatter(X_2d_rescaled[idx, 0], X_2d_rescaled[idx, 1], label='{}'.format(canton), marker='x')


# Labels and legend
plt.legend()
plt.xlabel('1st component')
plt.ylabel('2nd component')
plt.show()
```

##### Findings

The most extrem, BS and AI, are marked with a full arrow. The second most extrem, GE and NW, are marked with a light arrow. The two groups (the two cantons the most expensive and the two canton the less expensive) are clearly at the extremities of the plot. Moreover, each canton is clearly on some line, in other words each canton is isolated from others and the differents points form an evolution structure for each canton.

An interpretation of the two axes could be the premium level through the cantons from left to right, and the evolution of premiums through the years on a vertical axis.

Note about the cost of the premiums: in this work appeared two analysis axes, the evolution of costs, or the difference between years and the difference between cantons (independently from the years). Both are not separated in this work but it's better to keep it in mind.

```python
# there are the alignment of features on the PCA 2 axes:
results_df = pd.DataFrame.from_items([
    ('variance', X_rescaled.var(axis=0)),
    ('1st component', pca.components_[0]),
    ('2nd component', pca.components_[1])
]).set_index(data_df.drop(['Premium'], axis=1).columns)

# Sort DataFrame by variance
results_df.sort_values('1st component', ascending=False).head(10)
```

```python
# let's plot the data on a biplot:

plt.figure(figsize=(18,10))

for canton in data_df.iloc[:, 19:]:
    idx = (data_df.loc[:, canton] == 1)

    # Plot their components
    if(canton == 'Canton_BS'):
        plt.scatter(X_2d_rescaled[idx, 0], X_2d_rescaled[idx, 1], label='{}'.format(canton), marker='^')
    elif(canton == 'Canton_GE'):
        plt.scatter(X_2d_rescaled[idx, 0], X_2d_rescaled[idx, 1], label='{}'.format(canton), marker='2')
    elif(canton == 'Canton_NW'):
        plt.scatter(X_2d_rescaled[idx, 0], X_2d_rescaled[idx, 1], label='{}'.format(canton), marker='1')
    elif(canton == 'Canton_AI'):
        plt.scatter(X_2d_rescaled[idx, 0], X_2d_rescaled[idx, 1], label='{}'.format(canton), marker='v')
    else:
        plt.scatter(X_2d_rescaled[idx, 0], X_2d_rescaled[idx, 1], label='{}'.format(canton), marker='x')

# Plot a few features using the weights in the loading vectors:
# (the three firsts of the 1st and the 2nd components - negative value are less aligned here)
for feature in ['Doctors_density', 'Specialized doctors density', 'Women',
               'Expectancy Men', 'Expectancy Men 65', 'Men']: # three first components of each column
    # Get weights
    weight1 = results_df.loc[feature, '1st component']*8 # rescale them
    weight2 = results_df.loc[feature, '2nd component']*8 # (8 is arbitrary)

    # plot them
    plt.arrow(
        0, 0, # Vector starts at (x,y)=(0,0)
        weight1, weight2, # ends at (x,y)=(weight1, weight2)
        color='black', width=0.1)
    plt.text(
        weight1*1.5, # x location (1.5 is arbitrary)
        weight2*1.5, # y location
        feature, # Feature name
        weight='bold', color='xkcd:scarlet')

# Labels and legend
plt.legend()
plt.xlabel('1st component')
plt.ylabel('2nd component')
plt.show()
```

### Model computation: Decision tree and Random forest

The goal is to identify which feature split the best the data and use it to make prediction on premiums 2017 with the following step:

- for each of the three dataset:
    - create and optimise a decision tree regressor
    - show graphviz
    - create and optimise a random forest regressor
    - compute 2017 predictions and compare with 2017 premiums

```python
# now we separate 2017 values which will be used as test set:

# dataset with data from FOPH:
X_FOPH_2017 = data_df[data_df.Year == 2017].drop(['Premium'], axis=1)
y_FOPH_2017 = data_df[data_df.Year == 2017].Premium
data_df.drop(data_df[data_df.Year == 2017].index, inplace=True)

# dataset with data from FOPH and FSO: 
X_FSO_2017 = data_census_OH[data_census_OH.Year == 2017].drop(['Premium'], axis=1)
y_FSO_2017 = data_census_OH[data_census_OH.Year == 2017].Premium
data_census_OH.drop(data_census_OH[data_census_OH.Year == 2017].index, inplace=True)

# dataset with benefits data:
X_benefits_2017 = data_benefits_OH[data_benefits_OH.Year == 2017].drop(['Premium'], axis=1)
y_benefits_2017 = data_benefits_OH[data_benefits_OH.Year == 2017].Premium
data_benefits_OH.drop(data_benefits_OH[data_benefits_OH.Year == 2017].index, inplace=True)
```

```python
# creating DT (decision tree) and RF (random forest) using regressor function instead of classifier.
# They will reduce the error by splits based on features:

from sklearn.model_selection import train_test_split

X_tr, X_valid, y_tr, y_valid = train_test_split(data_df.drop(['Premium'], axis=1), data_df.Premium, 
                                                train_size=0.8, test_size=0.2, random_state=0)

print(X_tr.shape, y_tr.shape)
print(X_valid.shape, y_valid.shape)

from sklearn.tree import DecisionTreeRegressor

train_scores = []
valid_scores = []
overfitting_scores = []

for i in np.arange(1, 20): # grid search with depth from 1 to 19
    # create a decision tree :
    dt = DecisionTreeRegressor(max_depth=i, random_state=0)

    # fit estimator :
    dt.fit(X_tr, y_tr)
    
    # score results
    train_score = dt.score(X_tr, y_tr)
    valid_score = dt.score(X_valid, y_valid)
    overfitting_score = train_score / valid_score - 1
    
    # store scores
    train_scores.append(train_score)
    valid_scores.append(valid_score)
    overfitting_scores.append(overfitting_score)
```

```python
# plot of the train/valid curves to see the best parameter:
plt.figure(figsize=(13, 6))
plt.plot(np.arange(1, 20), train_scores, label='train')
plt.plot(np.arange(1, 20), valid_scores, label='valid')
plt.title('Train and validation curves')
plt.xticks(np.arange(1, 20))
plt.xlabel('Decision tree depth')
plt.ylabel('Model score')
plt.legend()
plt.show()
```

```python
# graphviz to visualize splitting:

from sklearn.tree import export_graphviz

dt = DecisionTreeRegressor(max_depth=3, random_state=0) # depth of 3 for a best graph visibility
dt.fit(X_tr, y_tr)

graph = export_graphviz(dt, out_file=None, feature_names=data_df.drop(['Premium'], axis=1).columns, 
                        class_names=['Premium'], filled=True, rounded=True, proportion=True)

import graphviz

# Display decision tree
graphviz.Source(graph)
```

##### Findings:

The first rule which reduce impurity mesure is a complementary benefits ratio under 4.2%. We saw before a correlation of 0.68 between complementary benfits ratio and premium. This can be seen as a mesure of the urbanisation (or "city life") of the canton. We saw before a link between premium and urbanisation or others social coverages and between urbanisation and this others social coverages. We see that about 66% of data follow this rule which have a mean premium of 3907. The mean premium of the other group is 5111.

Note about the complementary benefits ratio: this is the part of the population which receive a complementary benefit. These benefits are given to people which receive an AHV/IV annuity which revenues don't cover the minimum living costs (2014: 2/3 AHV annuitant, or people over 65 and 1/3 IV annuitant, or disable people under 65). We assume that in urbanisation area, people easily ask help and life is more expensive.

The second level of the tree show two differents features, the Year and the general practitioners density. Year make sense here because the evolution of premiums over it isn't regular, as seen before there's a rush around 2008 which makes premiums before this year significantly different than after this year. On the other side, general practitioners density make sense too, for the reasons explained before. It's interesting to see that this density is more critical in cantons with more complementary benefits ratio, or urbanisation.

```python
# validation seems to be good and stable since a depth of 11.
# Now we will use GridSearchCV, which will find the best parameters (depth and estimators numbers).
# This use cross-validation to optimize results:

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

train_scores = []
valid_scores = []
overfitting_scores = []

gs_rf = GridSearchCV(
    estimator=RandomForestRegressor(),
    param_grid={
        'max_depth': range(5,15),
        'n_estimators': (10, 50, 100, 150, 200)
    },
    cv=5, scoring='neg_mean_squared_error', n_jobs=-1)

gs_rf.fit(X_tr, y_tr)
best_params = gs_rf.best_params_

rfr = RandomForestRegressor(max_depth=best_params["max_depth"], n_estimators=best_params["n_estimators"],
                            random_state=False, verbose=False)# Perform K-Fold CV
rfr.fit(X_tr, y_tr)
rfr.score(X_tr, y_tr), rfr.score(X_valid, y_valid)
```

```python
# the best parameters found by the grid search are:
best_params
```

```python
# let's predict the 2017 premiums on basis of data:

y_pred = rfr.predict(X_FOPH_2017)

# plot of the predictions and the real 2017 premiums. The red line is the equality between both.
# A scatter with premium 2016 is shown for comparison:
x = np.linspace(4000, 7500)

plt.figure(figsize=(12, 7))
plt.scatter(y_FOPH_2017, y_pred, label='2017')
plt.scatter(data_df[data_df.Year == 2016].Premium.values, y_pred, label='2016')
plt.plot(x, x, color='red', label='equality between prediction and real')
plt.xlabel('Premiums')
plt.ylabel('Predictions 2017')
plt.legend()
plt.show()

# detail of the data:
rf_df = pd.DataFrame({'Canton': data[data.Year == 2017].Canton,
                      'Premiums 2017': y_FOPH_2017, 
                      'Predictions 2017': y_pred,
                      'Delta': y_FOPH_2017 - y_pred, 
                      'Premiums 2016': data_df[data_df.Year == 2016].Premium.values})
rf_df
```

### Models comparison: baseline and RMSE

##### Let's try to predict 2017 premiums from each dataset. But first, we have to find a baseline (text from proposal):

My understanding of the baseline is a landmark which compares the result of a simple model with our complicated model (with the hope that our complicated model is more precise than the simple model). Assuming that normal premiums behavior is rising (like prices, salaries, etc.), I prefer making an interpolating of the previous years, calculated as follow (example 2012-2016):

"$y = ax + b <=> y' = a(x+1) + b$ (where y' is the premium on next year) $<=> y' = ax + b + a <=> y' = y + a$

"and $y = ax + b <=> a = (y - b) / x => a = [premium 2016 - premium 2012]/[4 years]$, per canton

The baseline will be the error caracterised by the difference between this extrapolated premiums and the real 2017 premiums.

```python
# following this, we have for the first dataset (data from the FOPH):

a_FOPH_2017 = pd.merge(data[data.Year == 2004][['Canton', 'Premium']], 
                       data[data.Year == 2016][['Canton', 'Premium']], on=['Canton'])
a_FOPH_2017['Delta'] = ( a_FOPH_2017.Premium_y - a_FOPH_2017.Premium_x ) / 12
a_FOPH_2017['Premium'] = ( a_FOPH_2017.Premium_y + a_FOPH_2017.Delta)
bl_FOPH = a_FOPH_2017.drop(['Premium_x', 'Premium_y', 'Delta'], 
                           axis=1).sort_values(by=['Canton']).reset_index(drop=True)
bl_FOPH # bl = baseline
```

```python
# the second dataset (census from FSO) with the FOPH data:

a_census_2017 = pd.merge(data_census[data_census.Year == 2010][['Canton', 'Premium']], 
                         data_census[data_census.Year == 2016][['Canton', 'Premium']], on=['Canton'])
a_census_2017['Delta'] = ( a_census_2017.Premium_y - a_census_2017.Premium_x ) / 6
a_census_2017['Premium'] = ( a_census_2017.Premium_y + a_census_2017.Delta)
bl_census = a_census_2017.drop(['Premium_x', 'Premium_y', 'Delta'], 
                               axis=1).sort_values(by=['Canton']).reset_index(drop=True)
bl_census 
```

```python
# and the third dataset (santésuisse data, the cost groups):

a_benefits_2017 = pd.merge(data_benefits[data_benefits.Year == 2009][['Canton', 'Premium']], 
                           data_benefits[data_benefits.Year == 2016][['Canton', 'Premium']], on=['Canton'])
a_benefits_2017['Delta'] = ( a_benefits_2017.Premium_y - a_benefits_2017.Premium_x ) / 7
a_benefits_2017['Premium'] = ( a_benefits_2017.Premium_y + a_benefits_2017.Delta)
bl_benefits = a_benefits_2017.drop(['Premium_x', 'Premium_y', 'Delta'], 
                                   axis=1).sort_values(by=['Canton']).reset_index(drop=True)
bl_benefits
```

```python
# now we will compare these baselines with the reals premiums and will use RMSE because we are not 
# concerned with outliers and the results are smaller than MSE:

premiums2017_data = format_premiums(pd.read_excel(os.path.join('data', 'premiums_data', 
                                                               '2017A.xls'))).replace({'Year': -1}, 2017)
premiums2017_data.Premium = premiums2017_data.Premium * 12
premiums2017_data = premiums2017_data.drop(['Year'], 
                                           axis=1).sort_values(by=['Canton']).reset_index(drop=True)

premiums2017_data
```

```python
# in the course, we compared a baseline to differents models, showing the precision of each model.
# Here we compare different datasets, which are different explication of premiums, with a model applied 
# on each dataset.

def RMSE(y, y_pred):
    return np.sqrt(np.mean((y - y_pred)**2))

# rmse baseline with linear projection vs 2017 premiums: 
rmse_bl_FOPH = RMSE(bl_FOPH.Premium, premiums2017_data.Premium) # range of years: 2004-2016
rmse_bl_census = RMSE(bl_census.Premium, premiums2017_data.Premium) # range of years: 2010-2016
rmse_bl_benefits = RMSE(bl_benefits.Premium, premiums2017_data.Premium) # range of years: 2009-2016

rmse_values = [rmse_bl_FOPH, rmse_bl_census, rmse_bl_benefits]
rmse_titles = ['FOPH', 'FOPH+census', 'Benefits data']

xcor = np.arange(len(rmse_values))
plt.bar(xcor, rmse_values)
plt.xticks(xcor, rmse_titles)
plt.ylabel('RMSE')
plt.show()
```

```python
# now, let's apply limear/Ridge/Lasso regressions with grid search. Linear regression fit a polynomial
# on the data, Ridge applies a regularisation term, avoiding betas to overfitting the model.
# Lasso goes a bit further because it removes unnecessary betas.
# As we have three set to execute, we will first functionarize the algorithm. This function will 
# take data as input and return train and validation score and the estimator as output:

def lin_reg(data_X, data_y, regularization_type): # regularization_type = [ linear, ridge, lasso ]
    
    X_tr, X_valid, y_tr, y_valid = train_test_split(data_X.values, data_y.values, train_size=0.8,
                                                        test_size=0.2, random_state=0)
    
    if regularization_type == 'linear':
        
        from sklearn.linear_model import LinearRegression
        
        lr = LinearRegression()
        
        lr.fit(X_tr, y_tr)
        
        return lr.score(X_tr, y_tr), lr.score(X_valid, y_valid), lr
    
    if regularization_type == 'ridge':
        
        from sklearn.linear_model import Ridge
        
        grid_cv = GridSearchCV(Ridge(), {
            'alpha': np.logspace(-5, 1, num=100)
        }, cv=3)
        
        grid_cv.fit(X_tr, y_tr)        
        
        return grid_cv.score(X_tr, y_tr), grid_cv.score(X_valid, y_valid), grid_cv

    if regularization_type == 'lasso':
        
        from sklearn.linear_model import Lasso
        
        grid_cv = GridSearchCV(Lasso(max_iter=100000), {
            'alpha': np.logspace(-3, 10, num=100)
        }, cv=3, n_jobs=-1)
        
        grid_cv.fit(X_tr, y_tr)
        
        return grid_cv.score(X_tr, y_tr), grid_cv.score(X_valid, y_valid), grid_cv
```

```python
# So we have:
    # three datasets: FOPH, FOPH+census and benefits
    # three models: linear regression, ridge regression and lasso regression
    # three steps: fit estimators, show train/test scores and plot the results
```

```python
# Linear regression
# the linear regression is a simple model which try to fit the data in an multiple linear equation 
# of type ^yi = w0 + w1xi1 + w2xi2 + ... + wpxip <=> ^y = Xw + w0

lr_1_train, lr_1_valid, lr_1 = lin_reg(data_df.drop(['Premium'], axis=1), data_df.Premium, 'linear')
lr_2_train, lr_2_valid, lr_2 = lin_reg(data_census_OH.drop(['Premium'], axis=1), 
                                       data_census_OH.Premium, 'linear')
lr_3_train, lr_3_valid, lr_3 = lin_reg(data_benefits_OH.drop(['Premium'], axis=1), 
                                       data_benefits_OH.Premium, 'linear')

pd.DataFrame({'train score': [lr_1_train, lr_2_train, lr_3_train],
              'test score': [lr_1_valid, lr_2_valid, lr_3_valid]},
              index=['FOPH data', 'FOPH + census data', 'Benefits data'])
```

## Conclusion

Three components are recurrent and seems to explain the difference between premiums from these datasets, they are the doctors densities, the women percentage and a more subtil feeling of "city life".

##### Doctors densities

Except at the beginnig of the work where there is a break in the data, the doctors densities are really recurrent as top explanation in variance of premiums. This is even visible in the numbers itselves because the far expensiver cantons have a far high density of doctors (BS and GE). One understand the urge to maintain the control of the number of independant doctors offices now debated at the parlament.

##### Women as components of costs

As said, solidarity between women and men is part of the health insurance. This is concretised by a unique premium between both genders. In theory. In practice, Helsana published a study on premiums with differences between women and men, because of the choice of the franchise, the insurance model and, as seen above, because they live more in expensive areas. The difference of percentage of women through the canton was the real surprise here.

The informations highlithed above need to be interpreted carefully and with wisdom. There is no need to start a witch hunt, it would be much more clever to be empathic with women. With the #metoo phenomena, we learnt a lot on women condition in modern socities. What is the relationship between women and their health or care provider? The aim of my work is to dig in the layers, to understand the mechanisms. Women as an important component of costs is the indication where to go deeper. We all have heard women complaining about her gynecologist but not men about their generalist. I even seen a letter which would never been sent to a man. I think there is a lot to do in this direction to understand health system as seen by women.

##### City life

Some converging indices appears along the work. The two more expensive cantons are cities-canton and on the national boundaries, high premiums are correlated with social help and complementary benefits (in cities one ask more easily institutional help, in campaign one is supported by surroundings), a high rate of atheism, unemployement and tertiary level educated people, and a low level of wooden area. Moreover these two canton allow only drugs to be provided through pharmacies. This is not the urbanity like indicated in the FSO data, it isn't really precise but there's something. Or perhaps somethings.

##### A word again

The topic is very large and I refused myself to explore some features to avoid making the work too big. For example, "Type of drug providing" could be the topic of a whole work by itself. Is it better when doctors provides drug or worse? For the premium level? and for the cares quality? I tried to cover the topic at best without being to much verbose.

Another constatation is the choice of the input dataset. Models gives results based on the inputs. Data without doctors densities would give others results, probably with weaker predictions, and the models would have not design it as explanation in the variance of premiums. Models are blind of what they didn't see.
